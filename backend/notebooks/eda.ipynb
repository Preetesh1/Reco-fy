```python
# Cell 1 — Load data
import sys
sys.path.append("../src")
from data_loader import load_all
ratings, movies, users, matrix, train, test = load_all()

# Cell 2 — Rating distribution
import matplotlib.pyplot as plt
ratings["rating"].value_counts().sort_index().plot(kind="bar", color="steelblue")
plt.title("Rating Distribution")
plt.xlabel("Stars"); plt.ylabel("Count")
plt.show()
# Key insight: most ratings are 3–4. Very few 1s. This is typical rating bias.

# Cell 3 — Ratings per user (long tail)
ratings_per_user = ratings.groupby("user_id")["rating"].count()
ratings_per_user.hist(bins=50, color="coral")
plt.title("Ratings Per User")
plt.xlabel("# of ratings"); plt.ylabel("# of users")
plt.show()
# Key insight: most users rate very few movies. A few rate hundreds.
# This "long tail" is the cold start problem you'll handle Day 4.

# Cell 4 — Most rated movies
top_movies = (ratings.groupby("movie_id")["rating"]
              .agg(["count", "mean"])
              .join(movies.set_index("movie_id")["title"])
              .sort_values("count", ascending=False)
              .head(10))
print(top_movies)
# You'll recognize these: Star Wars, Fargo, Pulp Fiction etc.

# Cell 5 — Sparsity visualization (sample 50x50)
import seaborn as sns
sample = matrix.iloc[:50, :50].replace(0, float("nan"))
plt.figure(figsize=(12, 8))
sns.heatmap(sample, cmap="YlOrRd", cbar=True, xticklabels=False, yticklabels=False)
plt.title("User-Item Matrix (50×50 sample) — white = unrated")
plt.show()
# This visually shows WHY collaborative filtering is hard — mostly empty.
```
